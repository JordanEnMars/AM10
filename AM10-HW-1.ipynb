### Problem 2

Langrange points (L-points) are orbital points in which the gravitational forces of the two large bodies (of masses $M_1$ and $M_2$) cancel out. If an object (e.g. a satellite) is placed at an L-point, the object will be in equilibrium relative to the center of mass of the large bodies (the equilibrium can be stable or unstable, depending on the particular L-point). There are five such points, labeled L1 to L5, all in the orbital plane of the two large bodies. L1, L2, and L3 are on the line through the centers of the two large bodies, and were discovered by Leonhard Euler a few years before Joseph-Louis Lagrange, who in 1772 discovered the remaining two.

If $M_2 << M_1$ then L1 is located at approximately distance $r$ from the smaller object (of mass $M_2$), given by the equation:

$${\displaystyle r\approx R \ {\sqrt[{3}]{\frac {M_{2}}{3M_{1}}}}}$$

a) Consider that the Earth orbits the Sun in a perfect circle with separation equal to $R_{SUN-EARTH}$ (use the same value that you used in the previous problem) and write simple python code to calculate the distance of the L1 point from the Earth and from the Sun. Print the numerical value of your result in scientific notation (use 3 decimal digits) and describe with appropriate text (do not forget to include the relevant unit).

b) Calculate the distance of the Earth-Moon L1-point from the Earth using the values of $R_{EARTH-MOON}$, $M_{EARTH}$, $M_{MOON}$ from the previous problem. Print the numerical value of your result in scientific notation (use 3 decimal digits) and describe with appropriate text (do not forget to include the relevant unit).

NOTE OF HISTORICAL INTEREST:

The Sun–Earth L1 point is suited for making observations of the Sun–Earth system. Objects at this point are never shadowed by Earth or the Moon and, if observing Earth, always view the sunlit hemisphere. The first mission of this type was the 1978 International Sun Earth Explorer 3 (ISEE-3) mission used as an interplanetary early warning storm monitor for solar disturbances. Since June 2015, DSCOVR has orbited the L1 point. Conversely it is also useful for space-based solar telescopes, because it provides an uninterrupted view of the Sun and any space weather (including the solar wind and coronal mass ejections) reaches L1 a few hours before Earth. Solar telescopes currently located around L1 include the Solar and Heliospheric Observatory and Advanced Composition Explorer. [source: Wikipedia]



```python
RES=1.5042e11
MS1=1.9885e30
ME2=5.9722e24
r=RES*np.cbrt(ME2/(3*MS1))
z=RES-r
print('The point L1 is','%4.3e' % r,'meters away from the Earth')
print('The point L1 is','%4.3e' % z,'meters away from the Sun')
```
