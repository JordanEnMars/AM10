## 1. Rectified linear unit (ReLU)

The Rectified Linear Unit is a function that is used as activation in neural networks; it returns 0 if it receives any negative input, or it returns that value back if the input is positive. ReLU is the most commonly used activation function in deep learning models.

\begin{equation}
ReLU(x) = \begin{cases}
    x & \text{if } x > 0, \\
    0 & \text{otherwise}.
\end{cases}
\end{equation}

**Exercise:** Write a Python function to calculate ReLU for a given input $x$. Can you do this in one line? 

````python
def relu(x):
    if x>0:
        return x
    
    else:
        return 0
````

**Exercise:** Create a 1D array and transform it by passing it via the ReLU.  When we want to create a numpy array but we do not know the number of values in advance, we can create a list, populate it and then cast it as an array.

````python

x=np.array([2,6,8,-1,-4,7,8])
relu_x=[]
for elem in x:
    relu_x.append(relu(elem))
relu_x=np.array(relu_x,dtype='int64')
relu_x
    
````

**Exercise:** Plot the ReLU for some range of values.  

````python
x=np.linspace(-10,10,100)
y=[]
for elem in x:
    y.append(relu(elem))
y=np.array(y)
fig,ax=plt.subplots(figsize=(6,3))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('ReLU')
plt.title('Hello ReLU');
````

### Leaky ReLUs 

In the [Wikipedia](https://en.wikipedia.org/wiki/Rectifier_(neural_networks)) page you may find variations of ReLU. 

**Exercise** Choose one, e.g. the "Leaky" ReLU which allows a small, positive gradient to pass when the unit is not active. Define and plot it.

\begin{equation}
Leaky\_ReLU(x) = \begin{cases}
    x & \text{if } x > 0, \\
    0.01x & \text{otherwise}.
\end{cases}
\end{equation}

````python
def leaky_relu(x):
    if x>0:
        return x
    else:
        return 0.01*x
x=np.array([2,6,8,-1,-4,7,8])
leaky_relu_x=[]
for elem in x:
    leaky_relu_x.append(leaky_relu(elem))
leaky_relu_x=np.array(leaky_relu_x,dtype='int64')
leaky_relu_x
x=np.linspace(-10,10,100)
y=[]
for elem in x:
    y.append(leaky_relu(elem))
y=np.array(y)
fig,ax=plt.subplots(figsize=(6,3))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('Leaky ReLU')
plt.title('Hello Leaky ReLU');
## see screenshot
    
````

A **lituus** is a spiral described by the following equations: 

\begin{equation}
x(t)=r(t)cos(t)\\
y(t)=r(t)sin(t) 
\end{equation}

where 

$$r(t)=\sqrt{\frac{a^2}{t}}$$

The word lituus originally meant a curved war-trumpet in the ancient Latin language. This Latin word continued in use through the 18th century as an alternative to the vernacular names of various musical instruments. Mathematically, it is the locus of the point P moving such that the area of a circular sector remains constant *([source](https://en.wikipedia.org/wiki/Lituus))*

### 3.2 The `butterfly` equation

\begin{equation}
x = r(t)\cos(t) \\
y = r(t)\sin(t)\\
r(t) = e^{\sin(t)} - 2\cos(4t) + \sin^5 \left(\frac{2t - \pi}{24} \right) 
\end{equation}

````python
def butterfly(t,a):
    x=(np.exp(np.sin(t)))-2*np.cos(4*t)+((np.sin((2*t-np.pi)/24))**5)*np.cos(t)
    y=(np.exp(np.sin(t)))-2*np.cos(4*t)+((np.sin((2*t-np.pi)/24))**5)*np.sin(t)
    
n=100
a=5
t=np.arange(0.1,n*np.pi,.1)
butterfly(t,a)
plt.plot(x,y)
plt.show()
## see solutions
````
