### 1.1 Gaussian

The Gaussian function is defined as follows: 
\begin{equation}
f\left(x\right) = (\frac{1}{\alpha\sqrt{2\pi}}) \cdot \exp(\frac{{ - (x-x_0)}^2 )}{{2\alpha^2}}) 
\end{equation}

In the context of a Normal distribution:

- $x_0$=Expected mean $\mu$ 
- $\alpha^2$=Variance $\sigma^{2}$

**Exercise**: Define a Python function that returns a Gaussian function. Then plot it for a certain range. Make sure you include the right parameters in the definition.

````python
def gauss1D(x,x0,alpha):
    afact=1.0/(np.sqrt(2.0*np.pi)*alpha)
    return afact*np.exp(-(x-x0)**2/(2.0*alpha**2))
                        
xg= np.linspace(-10,10,100)
yg= gauss1D(xg,3.0,1.0)
plt.show(xg,yg)
plt.show()
````

### 1.2 `Poisson`

The **Poisson** function is used to count the discreet probability distribution of independent events occurring at a rate $\lambda$. One example is cars passing through a toll booth. Note that variable $x$ is an integer.

\begin{equation}
P_{\lambda}\left(x\right) = \frac{{e^{ - \lambda } \lambda ^x }}{{x!}} \quad
x = 0,1,2,...
\end{equation}

## 2. Beyond Functions: Parametric Equations for curves

A parametric equation defines a group of quantities as functions of one or more independent variables called parameters. Parametric equations are commonly used to express the coordinates of the points that make up a geometric object such as a curve or surface, in which case the equations are collectively called a parametric representation or parameterization of the object.
For example, the equations

$$x=\cos(⁡t), \quad	y=\sin⁡(t)$$

form a parametric representation of the unit circle, where t is the parameter: A point (x, y) is on the unit circle if and only if there is a value of t such that these two equations generate that point *(source:[Wikipedia](https://en.wikipedia.org/wiki/Parametric_equation))*.
