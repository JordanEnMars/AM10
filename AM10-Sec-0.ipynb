***Variables*** is an important way to specify data values in python.

***Example:***

```python
# Assigning strings to variables 

x = "Hello"
y = "World"
print(x)
print(y)
print(x, y)
print(x, y, "!")
print(x, y + "!")
```

```python
# Assigning numerical values to variables

num1 = 5
num2 = 7

print(num1 + num2)
```

**Naming variables in python**

The rules:
    
- You can use lowercase letters (a through z)
- You can use UPPERCASE letters (A through Z)
- You can use digits (0 through 9)
- You can use the underscore (_)
- Names are case sensitive (boston, Boston, and BOSTON are different names)
- Names must begin with a letter or an underscore, not a digit
- Names that begin with an underscore are treated specially
- Names cannot be one of python's reserved words. 

#### Python's reserved words

### Learning about ***Team Activities***
    
In AM10 Lectures and Labs students will work together in certain team activities. Each student participating will get assigned to a ***team***. Teams (of up to 6 students) work together in Lectures and Labs (but not in Sections). The team work (team activity) is structured as follows:

a) ***Working as Team on In-Class Problems***: During the ~10-15 min period of solving the In-Class problems (in the beginning of each Lecture), students will answer the questions/problems both as individuals and as a team. They will get a combined grade of their individual and their teamwork effort. A demo has been presented in the first class (Tue. 9/3 class, which has been recorded).

b) ***Working as Team in Lectures and Labs***: In each Lecture and Lab, there’ll be a couple of selected exercises and problems, for teams to work on (“team activities”), with allocated time of about ten minutes for each (for the Lectures, the In-Class problems team activity may be considered as one of them). Team members will get linked to the respective Breakout Rooms in order to work together. After the allocated time, team members will get back to class (main room).
In Breakout Rooms students will not (necessarily) write code, but they discuss the approach (solution strategy), difficulties encountered etc. When the teams come back to the main room (class), one team will describe their approach, possible problems encountered etc. Afterwards, the instructor will proceed with addressing the challenges and working on the solution.
